```python
import pandas as pd
```

# **_CREATE CSV CONNECTIONS_**

```python
# Fast food restaurants across US
fast_food = "Resources/Fast_Food_Restaurants_US.csv"
fast_food_df = pd.read_csv(fast_food)
fast_food_df
```

```python
# 18,000 Vegan & Vegetarian Restaurants in the US
vegan = "Resources/vegetarian_restaurants_US_datafiniti_vegetarian_restaurants_US_datafiniti.csv"
vegan_df = pd.read_csv(vegan)
vegan_df
```

# _Fast Food Locations DF_ Cleaning / Transformation

```python
# Create Copy of original CSV
fastfooddf = fast_food_df.copy()

# Find any duplicated rows in Fast Food df
fastfooddf = fastfooddf.drop_duplicates(['name', 'address', 'latitude', 'longitude'])
fastfooddf = fastfooddf.reset_index()
```

```python
# Remove Unnamed: 0 column as it is another index column
fastfooddf = fastfooddf.drop(columns=["Unnamed: 0"])

# Reorganize columns
fastfooddf = fastfooddf[["name", "categories", "address","city", "postalCode", "province", "country", "latitude", "longitude", "websites"]]

# Rename postalCode to postal_code
fastfooddf.rename(columns={'postalCode':'postal_code'}, inplace=True)
fastfooddf
```

# _Vegan Restaurants DF_ Cleaning / Transformation

```python
# Create Copy of original CSV and listed the columns
vegandf = vegan_df.copy()

# Find any duplicates in Vegan df
vegandf = vegandf.drop_duplicates(['name', 'address', 'lat', 'long'])
vegandf = vegandf.reset_index()

# REMOVE unwanted/unneeded columns by # the wanted columns
vegandf = vegandf.drop(columns=[
#address
#categories
#city
'claimed',
#country
#cuisines
'dateOpened',
'dateUpdated',
'descriptions',
'facebookPageURL',
'features',
'hours',
'images',
'isClosed',
'key',
#lat
'languages',
#long
'menus',
'menuURL',
#name
'paymentTypes',
'phones',
#postalCode
'priceRange',
#province
'sic',
'twitter'
#websites
])

# Reorganize columns
vegandf = vegandf[['name', 'categories', 'cuisines', 'address', 'postalCode', 'city', 'province',
                   'country','lat','long','websites']]

# Rename postalCode to postal_code
vegandf.rename(columns={'postalCode':'postal_code', 'lat':'latitude', 'long':'longitude'}, inplace=True)
vegandf
```

```python
# Removing [] and "" from the postalCode columns
vegandf['postal_code'] = vegandf['postal_code'].str.strip('[]')
vegandf['postal_code'] = vegandf['postal_code'].str.strip('""')

# Removing Brackets and " from the categories columns and replace , with /
# There are many / replacements inorder to make correct spacings for each category listed.
vegandf['categories'] = vegandf['categories'].str.strip('[]')
vegandf['categories'] = vegandf['categories'].str.strip('""')
vegandf['categories'] = vegandf['categories'].str.replace('","', ',')
vegandf['categories'] = vegandf['categories'].str.replace(',', '/')
vegandf['categories'] = vegandf['categories'].str.replace(' / ', '/')
vegandf['categories'] = vegandf['categories'].str.replace('/ ', '/')
vegandf['categories'] = vegandf['categories'].str.replace('/', ' / ')
vegandf.head()
```

# Save the dataframes into a new csv

```python
# Save Vegan DF
vegandf.to_csv("Output/vegan_restaurants_cleaned.csv")

# Save Fast Food DF
fastfooddf.to_csv("Output/fast_food_cleaned.csv")
```
