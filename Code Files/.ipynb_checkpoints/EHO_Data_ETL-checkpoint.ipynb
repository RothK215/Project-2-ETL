# **Read in the LA violation CSV**

#  **Clean the LA violation CSV**

```python
#clean LA violation data
# Create Copy of original CSV
laviolationdatadf = la_violation_data_df.copy()

# Find any duplicated rows in Fast Food df
laviolationdatadf = laviolationdatadf.drop_duplicates()

## drop NaN values
laviolationdatadf = laviolationdatadf.dropna() # removes 175 rows

# remove hours and seconds from the date
laviolationdatadf['activity_date'] = pd.to_datetime(laviolationdatadf['activity_date'])

# remove columns
laviolationdatadf = laviolationdatadf[['serial_number', 'activity_date', 'facility_name', 'violation_code', 
                                       'violation_description', 'violation_status', 'facility_id']]


laviolationdatadf
```

# **Clean the LA inspection CSV**

```python
#clean LA inspection data
# Create Copy of original CSV
ladatadf = la_data_df.copy()

# Find any duplicated rows in Fast Food df
ladatadf = ladatadf.drop_duplicates()

## drop NaN values
ladatadf = ladatadf.dropna() # only removes 3 rows

# remove hours and seconds from the date
ladatadf['activity_date'] = pd.to_datetime(ladatadf['activity_date'])

# remove columns
ladatadf = ladatadf[['serial_number', 'activity_date', 'facility_name', 'facility_address', 
                     'facility_city', 'facility_state', 'facility_zip', 'facility_id', 'grade', 
                     'score', 'program_status', 'record_id','service_description']]


ladatadf
```

# **Clean the Chicago inspection CSV**

```python
#clean chicago data
# Create Copy of original CSV
chicagodatadf = chicago_data_df.copy()

# Find any duplicated rows in chicago df
chicagodatadf = chicagodatadf.drop_duplicates()

## drop NaN values but we might want to keep them because they exist in the violations column
# chicagodatadf = chicagodatadf.dropna()
# chicagodatadf = chicagodatadf.replace(np.nan, '', regex=True) ## makes the NaN values blank

# fix zip code and license number format
chicagodatadf['Zip'] = chicagodatadf['Zip'].fillna(0)
chicagodatadf['Zip'] = chicagodatadf['Zip'].astype(int)
chicagodatadf['License #'] = chicagodatadf['License #'].fillna(0)
chicagodatadf['License #'] = chicagodatadf['License #'].astype(int)


# remove columns
chicagodatadf = chicagodatadf[['Inspection ID', 'Inspection Date', 'DBA Name', 'License #', 'Address',
                               'City', 'State', 'Zip', 'Latitude',
                             'Longitude', 'Inspection Type', 'Results', 'Violations']]

chicagodatadf = chicagodatadf.rename(columns={'Inspection ID': 'inspection_id', 
                                              'Inspection Date' : 'inspection_date', 
                                              'DBA Name' : 'dba_name', 
                                              'License #': 'license_number', 
                                              'Address' : 'address',
                                              'City' : 'city',
                                              'State' : 'state', 
                                              'Zip': 'zip', 
                                              'Latitude' : 'latitude',
                                              'Longitude' : 'longitude', 
                                              'Inspection Type': 'inspection_type', 
                                              'Results' : 'results', 
                                              'Violations' : 'violations'})

# remove hours and seconds from the date
chicagodatadf['inspection_date'] = pd.to_datetime(chicagodatadf['inspection_date'])


chicagodatadf
```
